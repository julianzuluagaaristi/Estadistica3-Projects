```python
import pandas as pd  # type: ignore
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore
import math
import random
```

```python

# Ruta al archivo Excel
ruta_archivo = "../datos-estadistica/datos.xlsx"

# Leer el archivo Excel
df = pd.read_excel(ruta_archivo)

# Mostrar todo el contenido del DataFrame
print(df)
```

```python
# Función para convertir porcentajes a float (elimina el símbolo %)
def convertir_porcentaje_a_numero(columna):
    return [float(valor.replace('%', '').strip()) for valor in columna]

# Función para calcular la media manualmente
def calcular_media(lista):
    return sum(lista) / len(lista)

# Función para calcular la mediana manualmente
def calcular_mediana(lista):
    lista_ordenada = sorted(lista)
    n = len(lista_ordenada)
    mitad = n // 2
    if n % 2 == 0:
        return (lista_ordenada[mitad - 1] + lista_ordenada[mitad]) / 2
    else:
        return lista_ordenada[mitad]

# Función para calcular la desviación estándar poblacional manualmente
def calcular_desviacion_estandar(lista):
    media = calcular_media(lista)
    varianza = sum((x - media) ** 2 for x in lista) / len(lista)
    return varianza ** 0.5

# Seleccionar solo las columnas numéricas y columnas que pueden contener porcentajes
columnas = df[['GDP (USD Billion)', 'Population (Millions)', 'Unemployment Rate (%)', 'Average Age', 'Women (%)', 'Men (%)', 'Budget (USD Billion)']]

# Nombres de las columnas a mostrar
columnas_nombres = ['GDP (USD Billion)', 'Population (Millions)', 'Unemployment Rate (%)', 'Average Age', 'Women (%)', 'Men (%)', 'Budget (USD Billion)']

# Recorrer las columnas numéricas y convertir valores si es necesario
for i, columna in enumerate(columnas.columns):
    # Convertir a lista
    lista = list(df[columna])
    
    # Si la columna contiene valores porcentuales, convertirlos a números
    if '%' in str(lista[0]):
        lista = convertir_porcentaje_a_numero(lista)
    
    # Cálculos estadísticos
    media = calcular_media(lista)
    mediana = calcular_mediana(lista)
    desviacion_estandar = calcular_desviacion_estandar(lista)

    # Mostrar resultados
    print(f"\n{columnas_nombres[i]}:")
    print(f"Media: {media}")
    print(f"Mediana: {mediana}")
    print(f"Desviación Estándar Poblacional: {desviacion_estandar}")
```

```python
def calculate_boxplot_values(data):
    sorted_data = np.sort(data)
    n = len(sorted_data)
    
    q1_pos = (n + 1) / 4
    q3_pos = 3 * (n + 1) / 4
    median_pos = (n + 1) / 2
    
    # Mediana
    if n % 2 == 0:
        median = (sorted_data[int(n / 2) - 1] + sorted_data[int(n / 2)]) / 2
    else:
        median = sorted_data[int(n / 2)]
    
    # Cuartil 1
    if q1_pos.is_integer():
        q1 = sorted_data[int(q1_pos) - 1]
    else:
        q1 = sorted_data[int(q1_pos) - 1] + (q1_pos - int(q1_pos)) * (sorted_data[int(q1_pos)] - sorted_data[int(q1_pos) - 1])
    
    # Cuartil 3
    if q3_pos.is_integer():
        q3 = sorted_data[int(q3_pos) - 1]
    else:
        q3 = sorted_data[int(q3_pos) - 1] + (q3_pos - int(q3_pos)) * (sorted_data[int(q3_pos)] - sorted_data[int(q3_pos) - 1])
    
    # Calcular IQR y límites
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Detectar outliers
    outliers = sorted_data[(sorted_data < lower_bound) | (sorted_data > upper_bound)]
    
    return q1, median, q3, lower_bound, upper_bound, outliers
```

```python
# Crear gráficos de boxplot manuales
def plot_boxplot(data, column_name):
    q1, median, q3, lower_bound, upper_bound, outliers = calculate_boxplot_values(data)
    
    fig, ax = plt.subplots(figsize=(3, 3))
    
    # Dibujar caja
    box = plt.Rectangle((0, q1), 1, q3 - q1, edgecolor='black', facecolor='lightblue')
    ax.add_patch(box)
    
    # Dibujar línea de mediana
    plt.plot([0, 1], [median, median], color='red', linestyle='--')
    
    # Dibujar bigotes
    plt.plot([0.5, 0.5], [lower_bound, q1], color='black')
    plt.plot([0.5, 0.5], [q3, upper_bound], color='black')
    
    # Dibujar los outliers
    if outliers:
        plt.scatter(np.full_like(outliers, 0.5), outliers, color='red', marker='o')
    
    # Configuración del gráfico de los boxplot
    plt.xlim(-0.5, 1.5)
    plt.ylim(min(data) - 1, max(data) + 1)
    plt.xticks([])
    plt.title(f'Boxplot Manual de {column_name}')
    
    plt.show()
```

```python
# Graficar boxplots para cada columna
for i, column in enumerate(columnas):
    plot_boxplot(column, columnas_nombres[i])
```
